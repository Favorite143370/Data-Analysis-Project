# Student lack of interest into higher education institution
## Problem Definintion
* **The recent gradual loss of interest in continuing higher education among SS3 students in Enugu South, Enugu State.**

## Data Collection
**Data was colleced through questionnaires issued to SS3 students in**
1. **Federal Government College, Enugu.**
2. **Queens Model Secondary School, Enugu.**

**Data Content**

* **Age:** age of the students
* **Gender:** Male/Female
* **School:** Federal Government College Enugu & Queensâ€™ Model Secondary School Enugu	
* **Parents:** If the student live with both parents or with single parent	
* **Sponsor:** The parents sponsoring their education {both or single}	
* **Occupation:** The occupation of the parent sponsoring the education
* **Family status:**  High class, Middle class or Lower class
* **Class:** The course he or she would like to study in higher institution {course}
* **Guidance/Counsellor:** If the student has visited their school Guidance and Counsellor {Yes/No}
* **Further education:** I f the students have the intentions of furthering their education after their secondary school
* **Loss of interest:** Factors that can causes loss of interest in furthering education {unemployment, financial constraints, peer influence, family obligations, poor educational background, invalid certificate i. e not making use of the certificate after acquiring it to get job.
* **Skills:**  skills they would like to acquire {Entrepreneurship program, vocational training, apprenticeship, technology skills}

**Total rows is one hundred and thirty one (131), and thirteen (13) columns**

### visualizing gender column

```python
# To get the count of genders
GENDER = Edu['GENDER'].value_counts().reset_index()

# visualizing the number of gender
sns.set_style("whitegrid")
ax = sns.barplot(x="GENDER", y='count', data=GENDER);
for container in ax.containers:
    ax.bar_label(container)
plt.title('Count of Gender');
```

**Observation:**
* Male has 66.
* Female has 65

```python
# To get the Number of sponsor occupation 
OCCUPATION = Edu['OCCUPATION_PARENT'].value_counts().reset_index()

# visualizing the number of occupation
sns.set_style("whitegrid")
label = ["business", "civil servant", "engineer", "banker", "teacher", "Dr", "lawyer", "nurse",
         "accountant", "lecturer", "surveyour", "seamstress","military", "baker ","pilot ","politician",
         "private investigator","online busienss","photographer"]
sns.barplot(x="OCCUPATION_PARENT", y='count', data=OCCUPATION)
plt.title('Count of Parent Occupation')
plt.xticks(rotation=75);
```

**Observation:**

* We discover that occupations of sponsor is more of business

```python
SPONSOR = Edu["SPONSOR"].value_counts().reset_index()

#visualizing to see the highest sponsor
sns.barplot(x='SPONSOR', y='count', data=SPONSOR)
plt.title('Count of Sponsor');
```

**OBSERVATION:**

* Both parents have the highest mark in sponsorship

```python
# To get the number of gender that has the highest age
plt.figure(figsize=(8,6))
sns.boxplot(x="GENDER", y="AGE", data=Edu)
plt.title('Boxplot of Gender Column');
```

**OBSERVATION:**
    
* Male students have the highest in age which is 19    

```python
AGE= Edu['AGE'].value_counts().reset_index()

label = [16,17,18,15,19,14]
ax = sns.barplot(x='AGE', y='count', data=AGE, order=label)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Count of Age");
```

**OBSERVATION:**
    
* We discover that student both male and female under the age of 16 have the highest count    

```python
ax = sns.countplot(x='AGE', data=Edu, hue='GENDER', order=label)
for container in ax.containers:
    ax.bar_label(container)
plt.title('Count of Age vs Gender');
```

**OBSERVATION:**

* This visualization shows us that student under age of 16 are more of females, 17 have more of males while 19 have only males and 14 have both male and femal equal

```python
sns.countplot(data=Edu, x='SKILLS', hue='GENDER')
plt.xticks(rotation=90)
plt.title('Count Skill vs Gender');
```

**OBSERVATION:**
    
* The male student have the intention of going into entrepreneurship program/business and apprenticeship while the females prefer vocational program. 

```python
sns.countplot(x='LOSS OF INTEREST', data=Edu, hue='FURTHER EDUCATION')
plt.xticks(rotation=90)
plt.title('Count Loss of Interest vs Further Education');
```

**OBSERVATION:**
    
* Those that want to futher their education are higher than those that doesn't 

```python
sns.countplot(x='LOSS OF INTEREST', data=Edu, hue='FAMILY_STATUS')
plt.xticks(rotation=90)
plt.title('Count Loss of Interest vs Family Status');
```

**OBSERVATION:**

```python
sns.countplot(x='LOSS OF INTEREST', data=Edu, hue='GENDER')
plt.xticks(rotation=90)
plt.title('Count Loss of Interest vs Gender');
```

**OBSERVATION:**

```python
sns.countplot(x='LOSS OF INTEREST', data=Edu, hue='SPONSOR')
plt.xticks(rotation=90)
plt.title('Count Loss of Interest vs Sponsor');
```

## Conclusion
   **From the observation, loss of interest in going into higher institution is influenced by some factors.**
  **These factors include financial constraints, Unemployment, Invalid certification, peer influence, family obligation, poor educational background, and intentions of marriage, but the two major factors are Financial constraints and Unemployment.**
   **The students in middle and lower classes are mostly affected.**
   **If these factors are erased, students will be more willing to further their education.**
   **Students can be encouraged to further their education by empowering them with skills immediately after leaving secondary school, and before entering the university. With these skills, such as Tech skills, Entrepreneurship and Business, the issue of financial constraints can be subsized while creating more jobs.**
