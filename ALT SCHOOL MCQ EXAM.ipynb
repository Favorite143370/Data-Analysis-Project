```python
Gender = Altschool['Gender'].value_counts()
Gender

# Plotting the bar chart
ax = Gender.plot(kind='bar', color=['brown', 'green'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

# Adding value labels on top of the bars
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    
plt.show()
```

```python
MathScore_per_group = Altschool.groupby('EthnicGroup')['MathScore'].mean().sort_values(ascending=False)
MathScore_per_group

# Create a bar plot
plt.figure(figsize=(8, 6))
ax = sns.barplot(x=MathScore_per_group.index, y=MathScore_per_group.values, palette='viridis')

# Add value labels on top of the bars
for index, value in enumerate(MathScore_per_group.values):
    ax.text(index, value, f"{value:.2f}", ha='center', va='bottom')

plt.xlabel('Ethnic Group')
plt.ylabel('Mean Math Score')
plt.title('Math Scores by Ethnic Group')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

```python
# Convert the column to string type to ensure all entries are treated as strings
Altschool['WklyStudyHours'] = Altschool['WklyStudyHours'].astype(str)

# Define a function to replace the specific value
def replace_study_hours(value):
    if value.strip() == "2023-05-10 00:00:00":
        return ">15"
    return value

# Apply the function to the column
Altschool['WklyStudyHours'] = Altschool['WklyStudyHours'].apply(replace_study_hours)
```

```python
Altschool
```

```python
# Create scatter plots for each subject
plt.figure(figsize=(12, 6))

# MathScore vs. Weekly Study Hours
plt.subplot(131)
sns.scatterplot(x='WklyStudyHours', y='MathScore', data=Altschool)
plt.title('MathScore vs. Weekly Study Hours')

# ReadingScore vs. Weekly Study Hours
plt.subplot(132)
sns.scatterplot(x='WklyStudyHours', y='ReadingScore', data=Altschool)
plt.title('ReadingScore vs. Weekly Study Hours')

# WritingScore vs. Weekly Study Hours
plt.subplot(133)
sns.scatterplot(x='WklyStudyHours', y='WritingScore', data=Altschool)
plt.title('WritingScore vs. Weekly Study Hours')

plt.tight_layout()
plt.show()
```

## Observation
* Overall, the scatter plots show that students' scores in math, reading, and writing can vary widely regardless of how many hours they study each week.

```python
# Convert the WklyStudyHours to a categorical type for better analysis
study_hours_mapping = {
    '<5': 0,
    '>10': 1,
    '>15': 2
}
Altschool['WklyStudyHours'] = Altschool['WklyStudyHours'].map(study_hours_mapping)

# Calculate the correlation matrix
correlation_matrix = Altschool[['WklyStudyHours', 'MathScore', 'ReadingScore', 'WritingScore']].corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation between Weekly Study Hours and Academic Performance')
plt.show()

# Analyzing the data
study_hours_summary = Altschool.groupby('WklyStudyHours')[['MathScore', 'ReadingScore', 'WritingScore']].describe()
```

## Observation:
* Students who study more tend to score higher in MathScore, ReadingScore, and WritingScore.
